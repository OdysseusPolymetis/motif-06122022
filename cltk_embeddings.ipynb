```python
from cltk import NLP
```

```python
cltk_lat = NLP(language="lat")
```

```python
cltk_doc = cltk_lat.analyze(text="Lorem ipsum dolor sit amet, consectetur adipiscing elit. ")
```

```python
for word in cltk_doc:
  print(word.embedding)
```

```python
from cltk.embeddings.embeddings import FastTextEmbeddings
```

```python
fasttext_obj = FastTextEmbeddings(iso_code="lat", silent=False)
```

```python
print(fasttext_obj.get_sims("minerva"))
```

```python
from cltk.embeddings.embeddings import Word2VecEmbeddings
```

```python
w2v_obj = Word2VecEmbeddings(iso_code="lat", silent=False)
```

```python
print(w2v_obj.get_sims("minerva"))
```
