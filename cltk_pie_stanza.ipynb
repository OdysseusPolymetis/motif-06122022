```python
from cltk import NLP
```

```python
cltk_lat = NLP(language="lat")
```

```python
cltk_doc = cltk_lat.analyze(text="Lorem ipsum dolor sit amet, consectetur adipiscing elit. ")
```

```python
for elem in cltk_doc:
  print(elem.string)
  print(elem.lemma)
  print(elem.upos)
  print("XXXXXXXXXXX")
```

```python
model_name = "lasla"
tagger = get_tagger(model_name, batch_size=256, device="cuda", model_path=None)
from pie_extended.models.lasla.imports import get_iterator_and_processor
for sentence_group in sentences:
    iterator, processor = get_iterator_and_processor()
    for elem in tagger.tag_str(sentence_group, iterator=iterator, processor=processor):
      print(elem)
```

```python
import stanza
```

```python
stanza_lat = stanza.Pipeline('la', processors='tokenize,lemma,pos,depparse')
```

```python
stanza_results = stanza_lat("Lorem ipsum dolor sit amet, consectetur adipiscing elit. ")
for sent in stanza_results.sentences:
  for word in sent.words:
    print(word.text)
    print(word.lemma)
    print(word.upos)
    print(word.feats)
    print("XXXXXXXXXXXX")
```
